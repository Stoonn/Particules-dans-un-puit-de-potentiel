###  1- Etablir les équations que verifie la fonction d'onde d'une particule quantique dans un puit de potentiel (fini) en fonction de sa position (différentes Zones)

Dans les années 1930, De Broglie a fait l'hypothèse de faire une analogie entre les électrons et les photons, ce qui permettrait de faire le lien entre l'énergie d'un photon : $E = h\times\nu$ et celle d'un corps massique $ E = m \times c^2$. 

Cela a permis d'associer à chaque corpuscule matériel une longueur d'onde $ \lambda = \frac{h}{mv}$, et donc une équation d'onde. Cette hypothèse a été vérifiée par l'expérience des fentes Young.

On peut donc écrire l'équation d'onde d'un système stationnaire de façon analogue à celle de la lumière: 
$$ \psi (x,y,z,t)=\psi(x,y,z)\phi(t)$$

A une dimension, on obtient : $$\psi(x,t) = \psi(x)\phi(t)$$

On peut montrer, via des équations qu'on verra au semestre prochain que la fonction d'onde doit vérifier cette équation :
$$i\hbar\frac{\partial\psi(x,t)}{\partial t}=E\psi(x,t)$$

Alors, pour un état stationaire de l'énergie la fonction d'onde $\psi$(x, t) peut s'écrire : $$\psi(x, t) = \psi(x)e^{-iEt/\hbar}$$

De plus, on sait que cette fonction est solution de l'équation de schrödinger : $$-\frac{\hbar^2}{2m}\frac{\partial^2\psi(x)}{\partial x^2} + [V(x)-E]\psi(x) = 0$$

Enfin, le potentiel $V(x)$ est défini tel que vu sur le schéma :
- si  $x< -a,\qquad V(x) = V_{0}$ 
- si  $-a < x < a,\ V(x) = 0$
- si $x >a,\qquad \quad V(x) = V_{0}$

**Pour la region 1**,

$x < -a$ &nbsp;&nbsp;&nbsp;donc : 
- $V(x) = V_{0}$
- $E < V_{0}$

En posant $k = \sqrt{\frac{2m(V_{0} - E)}{\hbar^{2}}}$, l'équation que vérfie $\psi$ est : $\frac{\partial^{2}\psi(x)}{\partial x^{2}} = k^{2}\psi(x)$

Les solutions de cette équations s'écrivent : $\psi_{1}(x) = Ae^{kx} + Be^{-kx}$

Or, on sait que la probabilité d'une telle onde est normalisée donc comme $k>0$, le terme qui explose lorsque $x \longrightarrow -\infty$ n'est pas physiquement valable et ainsi $B = 0$.

Les solutions deviennent : 
$$\psi_{1}(x) = Ae^{kx}$$





```python

# Paramètres
A = 1
k = 1
# bord du puits
a = 2  

# Zone 1 : x < -a
x = np.linspace(-8, -a, 500)
psi = A * np.exp(k * x)

plt.plot(x, psi, color='blue', label=r'$\psi_1(x) = Ae^{kx}$')
plt.axvline(x=-a, color='red', linestyle='--', label='x = -a')
plt.xlabel('x')
plt.ylabel(r'$\psi_1(x)$')
plt.title(r'Fonction d \' onde en Zone 1')
plt.legend()
plt.show()
```

**Pour la region 2**,

$-a <x < a$ &nbsp;&nbsp;&nbsp;donc : 
- $V(x) = 0$

En posant $q = \sqrt{\frac{2mE}{\hbar^{2}}}$, l'équation que vérfie $\psi$ est : $\frac{\partial^{2}\psi(x)}{\partial x^{2}} = -q^{2}\psi(x)$

Les solutions de cette équations s'écrivent : $$\psi_{2}(x) = C\cos(qx) + D\sin(qx)$$





```python

# Paramètres
C = 1
D = 1
q = 1
# bords du puits
a = 2  

# Zone 2 : -a < x < a
x = np.linspace(-a, a, 500)
psi = C * np.cos(q * x) + D * np.sin(q * x)

plt.plot(x, psi, color='green', label=r'$\psi_2(x) = C\cos(qx) + D\sin(qx)$')
plt.axvline(x=-a, color='red', linestyle='--', label='x = -a')
plt.axvline(x=a, color='red', linestyle='--', label='x = a')
plt.xlabel('x')
plt.ylabel(r'$\psi_2(x)$')
plt.title(r'Fonction d\'onde en Zone 2')
plt.legend()
plt.show()
```

**Pour la region 3**,

$x > a$ &nbsp;&nbsp;&nbsp;donc : 
- $V(x) = V_{0}$
- $E < V_{0}$

En posant $k = \sqrt{\frac{2m(V_{0} - E)}{\hbar^{2}}}$, l'équation que vérfie $\psi$ est : $\frac{\partial^{2}\psi(x)}{\partial x^{2}} = k^{2}\psi(x)$

Les solutions de cette équations s'écrivent : $\psi_{3}(x) = Fe^{kx} + Ge^{-kx}$

Or, on sait que la probabilité d'une telle onde est normalisée donc comme $k>0$, le terme qui explose lorsque $x \longrightarrow +\infty$ n'est pas physiquement valable et ainsi $F = 0$.

Les solutions deviennent : 
$$\psi_{3}(x) = Ge^{-kx}$$





```python

# Paramètres
G = 1
k = 1
# bord du puits
a = 2  

# Zone 3 : x > a
x = np.linspace(a, 8, 500)
psi = G * np.exp(-k * x)

plt.plot(x, psi, color='blue', label=r'$\psi_3(x) = Ge^{-kx}$')
plt.axvline(x=a, color='red', linestyle='--', label='x = a')
plt.xlabel('x')
plt.ylabel(r'$\psi_3(x)$')
plt.title(r'Fonction d \' onde en Zone 3')
plt.legend()
plt.show()
```

### 2- Etablissement des équations de raccordements ($x = a$)

En $x = -a$, la fonction d'onde et sa dérivée sont continues 

$\psi_{2}(-a) = \psi_{1}(-a)$ et : $\psi_{2}'(-a) = \psi_{1}'(-a)$ 

$\Leftrightarrow \begin{cases}
C\cos(qa)-D\sin(qa) = A\exp(-ka) \\ 
q(C\sin(qa)+D\cos(qa)) = kA\exp(-ka)
\end{cases}$

On injecte la première equation dans la deuxième: 

$\Rightarrow
q(C\sin(qa) + D\cos(qa)) = k(C\cos(qa) - D\sin(qa)) \\
\Rightarrow
C(q\sin(qa) + kcos(qa)) + D(q\cos(qa) - k\sin(qa)) = 0$

On obtient une solution impaire vérifiant $$\color{green}{-\frac{q}{\tan(qa)} = k}$$ et une solution paire vérifiant $$\color{green}{q\tan(qa) = k}$$

Ces 2 équations nous permettrons de trouver les énergies des particules coincées dans le puit.

Nous obtenons alors, pour les solutions paires des fonctions d'onde de la forme : 
$$\begin{cases}
x<-a : \psi_1(x) = A\exp(kx) \\
x \in [-a;a] : \psi_2(x) = Bcos(qx) \\
x>a : \psi_3(x) = A\exp(-kx)
\end{cases}$$


En x = -a, on a $\psi_1(-a) = \psi_2(-a) \Leftrightarrow A = B\cos(qa)\exp(ka)$

La seconde équation qui nous permettra de determiner B est la normalisation de la fonction d'onde. La fonction d'onde en elle même n'a pas de sens physique, c'est le module de son carré qui nous donne la probabilité de présence de la particule à un endroit donné. 

C'est une probabilité, donc intégrée sur tout l'espace, elle renvoie 1 :  

$$\int_{-\infty}^{+\infty} |\psi(x)|^2\,dx = 1$$

$$ 2\int_{-\infty}^{-a} A^2\exp(2kx)\,dx +\int_{-a}^{a} B^2\cos(qx)^2,dx = 1$$

$$\Leftrightarrow \frac{A^2\exp(-2ka)}{k} + B^2(a+\frac{\sin(2qa)}{2q}) = 1$$

On obtient le système suivant: 

$$\begin{cases}
A = B\cos(qa)\exp(ka) \\
\frac{A^2\exp(-2ka)}{k} + B^2(a+\frac{\sin(2qa)}{2q}) = 1
\end{cases}$$

dont les solutions sont : 
$$\begin{cases}
B = \frac{1}{\sqrt{a + \frac{sin(2qa)}{2q} + \frac{\cos(qa)^2}{k}}} \\
A = \frac{\cos(qa)\exp(ka)}{\sqrt{a + \frac{sin(2qa)}{2q} + \frac{\cos(qa)^2}{k}}}
\end{cases}$$

De manière analogue, les coefficient des fonctions d'ondes pour des solutions impaires sont: 

$$\begin{cases} 
B = \frac{1}{\sqrt{a - \frac{\sin(2qa)}{2q} + \frac{\sin^2(qa)}{k}}} \\
A = \frac{\exp(ka)\sin(qa)}{\sqrt{a - \frac{\sin(2qa)}{2q} + \frac{\sin^2(qa)}{k}}}
\end{cases}$$

Nous pouvons donc maintenant exprimer les fonctions d'ondes ! 

Attention il y a quand même une subtilité, chaque fonction d'onde dépend de q et k, et q et k dépendent de l'énergie de la particule. Alors, il ne nous reste plus qu'à déterminer les énergies quantifiées.




### 3- Calculs des énergies propres (où presque)

Reprenons ce que nous savons sur l'énergie: 

-> En reprenant l'équation de schrödinger :

On se rend compte que (E-V) doit être positif pour que l'équation ait des solutions réelles. 

De plus, E intervient dans le calcul de q : $q = \sqrt{\frac{2mE}{\hbar^{2}}}$ et dans le calcul de k : $k = \sqrt{\frac{2m(V_{0} - E)}{\hbar^{2}}}$

Et comme on l'a vu plus haut, les équations d'onde doivent verifier une équation : soit $-\frac{q}{\tan(qa)} = k$ si $\psi$ est impaire soit $q\tan(qa) = k$ si $\psi$ est paire.

On va donc essayer d'approcher numeriquement E pour que ces équations soient vérifiées. On va avoir besoin de la MAO car ces équations ne sont pas résolubles algébriquement. 

On pose $E_p = k + \frac{q}{\tan(qa)}$ et $E_i = k - q\tan(qa)$

Comme on travaille avec des valeurs de $V_0$ assez faible de d'ordre de grandeur $10^1$, on peut se permettre de travailler en évaluant les 2 fonctions sur l'intervalle [0,$V_0$] sans que ça ne fasse bruler aucun processeur (ce qui simplifie grandement la fonction d'évaluations). On va chercher dans cet intervalle les valeurs de E tel que $E_p(E) \approx 0$ et $E_i(E) \approx 0$. Ces énergies seront celles des particules présentent et "bloquées" dans le puit de potentiel de hauteur $V_0$.

Il y a malheureusement une dernière subtilité... Les fonctions $E_p$ et $E_i$ ne sont pas continues. Il faut donc ajouter au code une condition qui estime la distance entre 2 évaluations consécutives des 2 fonctions et verifier qu'elle est inférieur à un certain seuil qu'on peut fixer assez haut car la fonction tangente évolue vite.

Nous avons finalement toutes la recette pour afficher la fonction d'onde d'une particule quantique dans un puit de potentiel, il ne reste plus qu'à choisir la largeur (2a) du puit et la hauteur (V_0) du puit dans la simulation "puit de potentiel"!

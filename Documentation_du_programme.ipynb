###  1- Structure du programme

L'objectif était de modéliser un puit de potentiel quantique sur une interface graphique modifiable.

Pour cela nous avons opté pour une interface graphique **Tkinter** et nous y avons implémenté notre figure.

Le programme est composé d'une classe nommée **InterfaceGraphique** dont la classe parente est la classe **Tk()** provenant de **Tkinter**. La classe **Tk()** permet de créer la fenêtre de notre interface graphique.

Nous avonc donc d'abord défini une méthode **__init__()** dans laquelle nous allons instancier **Tk()** et modéliser notre fenêtre.

*Exemple ci-dessous:*

```python
import tkinter as tk

class ExempleInterface(tk.Tk):
    def __init__(self):

        # intanciation de Tk()
        tk.Tk.__init__(self)

        # paramétrisation de la fenêtre
        self.grid_rowconfigure(0, weight=2)
        self.grid_rowconfigure(1, weight=10)
        self.grid_columnconfigure(0, weight=1)
        self.grid_columnconfigure(1, weight=7)

        self.control_frame = tk.Frame(self, bg="#0c1018",
                                      relief="sunken", bd=1)
        self.control_frame.grid(row=1, column=0, sticky='nsew')


Interface = ExempleInterface()
Interface.mainloop()
```

Ensuite nous avons créé une Figure que nous avons implémenté dans **Tkinter**.

Pour cela nous avons tout d'abord importé la classe **Figure()** de **matplotlib.figure** dont l'instanciation nous permet de créer le conteneur vide pour notre figure. De plus la classe **Figure()** possède une méthode appelée **add_subplot()** qui permet d'instancier une autre classe appelée **Axes()** qui elle, permet de créer notre figure.

Afin d'implémenter cette figure dans **Tkinter** nous avons dû appeler une classe nommée **FigureCanvasTkAgg** provenant de **matplotlib.backends.backend_tkagg**, concrètement l'instanciation de cette classe va nous permettre de créer un objet **Tkinter** pouvant contenir notre figure.

Notons que l'utilisation de la méthode **get_tk_widget()** appartenant à **FigureCanvasTkAgg** nous permet de configurer le contenant de la figure au sein de Tkinter et ainsi son emplacement.

*Exemple ci-dessous:*

```python
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure

class ExempleFigure(tk.Tk):
    def __init__(self):

        # intanciation de Tk()
        tk.Tk.__init__(self)

        # paramétrisation de la fenêtre
        self.grid_rowconfigure(0, weight=2)
        self.grid_rowconfigure(1, weight=10)
        self.grid_columnconfigure(0, weight=1)
        self.grid_columnconfigure(1, weight=7)

        self.control_frame = tk.Frame(self, bg="#0c1018",
                                      relief="sunken", bd=1)
        self.control_frame.grid(row=1, column=0, sticky='nsew')

        # figure matplotlib
        self.fig = Figure()
        self.ax = self.fig.add_subplot(111)

        # insertion dans subplot
        self.canvas = FigureCanvasTkAgg(self.fig)
        self.canvas.get_tk_widget().grid(row=1,
                                         rowspan=40, column=1, sticky='nsew')


Exemple = ExempleFigure()
Exemple.mainloop()
```

Enfin notre programme possède de nombreuse méthode dont la majorité sont dédiées au traçage des fonction sur notre figure, les méthodes restantes permettent de changer la nature des fonctions à tracer ou bien de les retracer en fonction de nouveaux paramètres.
